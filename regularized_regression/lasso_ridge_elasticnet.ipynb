```python

def adj_r2(X, y, model):
    r2 = model.score(X, y)
    n = X.shape[0]
    p = X.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    
    return adjusted_r2
```

```python
adj_r2(X_train, y_train, lr), adj_r2(X_test, y_test, lr)
```
